```python
import pandas as pd

# Load the CSV file from the URL
table2 = pd.read_csv('https://raw.githubusercontent.com/ABHIBRO123/CAPSTONE-PROJECT/main/table2.csv')

# Display the first few rows of the dataframe
print(table2.head())
```

```python
import pandas as pd

# Load the CSV file from the URL
table3 = pd.read_csv('https://raw.githubusercontent.com/ABHIBRO123/CAPSTONE-PROJECT/main/table3.csv')

# Display the first few rows of the dataframe
print(table3.head())
```

```python
Now 3.	Merge the dataframes on SLNo (assuming SLNo is the common key):
```

```python
merged_df = pd.merge(table1, table2, on='SLNo')
```

```python
merged_df = pd.merge(merged_df, table3, on='SLNo')
```

```python
. Univariate Analysis:---
```

```python
# Distribution of property prices
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['Buy_total_price'], kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already loaded your dataframe merged_df
# merged_df = ...

# Distribution of property prices
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['Buy_total_price'], kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of carpet area
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['Carpet_area'], kde=True)
plt.title('Distribution of Carpet Area')
plt.xlabel('Carpet Area (sqft)')
plt.ylabel('Frequency')
plt.show()
```

```python
. Bivariate Analysis:-----
```

```python
#Price per square foot vs. Carpet area
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='Carpet_area', y='Price_per_square_feet', hue='City')
plt.title('Price per Square Foot vs. Carpet Area')
plt.xlabel('Carpet Area (sqft)')
plt.ylabel('Price per Square Foot')
plt.show()
```

```python
# Booking amount vs. Buy total price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x='Booking_amount', y='Buy_total_price', hue='Furnished_status')
plt.title('Booking Amount vs. Buy Total Price')
plt.xlabel('Booking Amount')
plt.ylabel('Total Price')
plt.show()
```

```python
CATOGERIAL ANALYSIS:---
```

```python
# Number of properties by city
plt.figure(figsize=(12, 8))
sns.countplot(data=merged_df, x='City', order=merged_df['City'].value_counts().index)
plt.title('Number of Properties by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Number of properties by furnished status
plt.figure(figsize=(10, 6))
sns.countplot(data=merged_df, x='Furnished_status')
plt.title('Number of Properties by Furnished Status')
plt.xlabel('Furnished Status')
plt.ylabel('Count')
plt.show()
```
